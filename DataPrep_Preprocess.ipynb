## Data Preparation
### We want to answer following questions from the data analysis which will help us build better recommendation
#### 1. Type of listings
#### 2. Price distribution with respect to room type
#### 3. Which area is expensive?
#### 4. How popular is airbnb(No.of reviews over the years)
#### 5. Which listings got the overall high rating score

```python
#Cobine name and description into single column- we may need it for content based similarity
df['content'] = df[['name', 'description']].astype(str).apply(lambda x: ' // '.join(x), axis = 1)
```

```python
df['content']
```

```python
#Filter out necessary columns for analysis
listing_df = df.filter(items=['id','content','neighbourhood_group_cleansed','latitude','longitude','bathrooms','bedrooms',
                        'room_type','price','minimum_nights','number_of_reviews','review_scores_rating'])
```

```python
listing_df
```

```python
listing_df.info()
```

#### The column price, which is the price of the listing, it contains the dollar sign ($)
#### remove the dollar symbol and convert the values as numeric values

```python
listing_df['price'] = listing_df['price'].str.replace("[$, ]", "").astype("float")
```

```python
listing_df['price']
```

```python
#Check for null values
listing_df.isna().sum()
```

```python
#Check for duplicates = If length of unique list id = No.of entries, there is no duplicate entry
len(df.id.unique())
```

```python
#Impute missing values for bedroom and bathroom with median and review_scores_rating with 0
listing_df.bathrooms.fillna(listing_df.bathrooms.median(),inplace=True)
listing_df.bedrooms.fillna(listing_df.bathrooms.median(),inplace=True)
listing_df.review_scores_rating.fillna(0,inplace=True)
```

```python
#Check for null values
listing_df.isna().sum()
```

```python
listing_df.to_csv('Clean_listings.csv',index=False)
```
