## Using LangDetect library 
Langdetect api does not parse the alphanumeric and numeric values, so we have to do exception handling as it throws errors when it encounters non-alpha characters.


```python
filtered_list = []
from langdetect import detect
for i in range(len(input_df)):
    try:
        language = detect(input_df.iloc[i].comments)
        if language=='en':
            filtered_list.append(input_df.iloc[i])
    except:
        print("This row throws an error:", input_df.iloc[i].comments + str(i))
```

### Creating a new dataframe with filtered list

```python
filtered_df = pd.DataFrame(filtered_list,columns=['listing_id','date','reviewer_id','comments'])
```

```python
filtered_df.to_csv('preprocessed_reviews.csv',index=False)
```
