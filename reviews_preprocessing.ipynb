```python
input_df = pd.read_csv('Clean_reviews.csv')
input_df.head()
```

```python
from langdetect import detect
for i in range(len(input_df)):
    try:
        language = detect(input_df.iloc[i].comments)
        if language is not 'en':
            input_df.drop(i,inplace=True)
    except:
        language = "error"
        print("This row throws an error:", input_df.iloc[i].comments)
        input_df.drop(i,inplace=True)
```
